```python
# Import required Python libraries:
# - requests: to make HTTP requests to the BEA API
# - pandas: to structure and clean the returned data
import requests, pandas as pd

# Your personal BEA API key (get one at https://apps.bea.gov/api/signup/)
KEY = "290EFCBB-FAEF-4B15-AED6-BBFB7B290205"

# Base URL for the BEA API
url = "https://apps.bea.gov/api/data/"

# Define the parameters for the API call
params = {
    "UserID": KEY,              # Your API key for authentication
    "method": "GetData",        # BEA API method to request actual data
    "datasetname": "Regional",  # Dataset = "Regional" (county/state/industry data)
    "TableName": "CAGDP2",      # Nominal GDP by county and industry (current dollars)
    "LineCode": "68",           # Industry = Educational services, health care, and social assistance (NAICS 61, 62)
    "Year": "2022",             # Year of data you want
    "GeoFIPS": "CA",            # Geography = California; returns all counties in CA
    "ResultFormat": "JSON",     # Response format (JSON is easiest to parse)
}
```

```python
# Make the API request with a timeout of 60 seconds
r = requests.get(url, params=params, timeout=60)

# Raise an exception if the HTTP request failed (e.g., bad URL or server issue)
r.raise_for_status()

# Convert the JSON text response to a Python dictionary
payload = r.json()

# Extract the actual data section of the API response
# The BEA JSON format wraps data under BEAAPI → Results → Data
data = payload["BEAAPI"]["Results"]["Data"]

# Create a pandas DataFrame from the JSON list of records
df = pd.DataFrame(data)
```

```python
df
```

```python
# Normalize all column names to lowercase for consistency
df.columns = [c.lower() for c in df.columns]
df
```

```python
# Keep only rows where GeoFIPS is a 5-digit number (i.e., counties, not state totals)
df = df[df["geofips"].str.len() == 5].copy()
df
```

```python
# Convert the GDP value (DataValue) from a string with commas (e.g., "1,234") to a numeric column
df["gdp_current_usd"] = pd.to_numeric(df["datavalue"].str.replace(",", ""), errors="coerce")
```

```python
# Rename columns for clarity:
# - geofips → county_fips (standard FIPS code)
# - timeperiod → year
df = df.rename(columns={"geofips":"county_fips", "timeperiod":"year", "geoname":"county_name"})
df
```

```python
# Convert the year column from string to integer
df["year"] = df["year"].astype(int)
```

```python
# Add a simple identifier for this combined sector (NAICS 61 & 62)
df["naics2_group"] = "61"
df
```
