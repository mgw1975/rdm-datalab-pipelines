```python
import pandas as pd

df = pd.read_csv("/Users/michaelwalker/Downloads/CAGDP2/CAGDP2__ALL_AREAS_2001_2023.csv", low_memory=False, encoding='latin1')
```

```python
df.head(10)
```

```python
#replaced by "cols" code on next line" df_tidy = df[["GeoFIPS", "Region", "IndustryClassification", "Description", "2022"]]
```

```python
cols = ["GeoFIPS", "LineCode", "IndustryClassification", "Description", "2022", "2021","2020"]
df_tidy = df.loc[:, cols].copy()
```

```python
df_tidy = df_tidy.rename(columns={"GeoFIPS":"state_county_fips_cd","LineCode":"line_cd", "IndustryClassification":"naics_sector_cd", "Description":"naics_sector_desc", "2022":"2022_gdp_num", "2021":"2021_gdp_num", "2020":"2020_gdp_num"})
```

```python
df_tidy['state_county_fips_cd'] = df_tidy['state_county_fips_cd'].str.replace('"', '')
```

```python
df_tidy['state_county_fips_cd'] = df_tidy['state_county_fips_cd'].str.replace(' ', '')
```

```python
df_tidy = df_tidy[df_tidy["state_county_fips_cd"].str.len() == 5].copy()
```

```python
mask = df_tidy["state_county_fips_cd"] != "00000"
```

```python
df_tidy2 = df_tidy[mask]
```

```python
df_tidy2
```

```python
mask2 = ((df_tidy2["2022_gdp_num"] != '(D)') & (df_tidy2["2022_gdp_num"] != '(NA)')) | ((df_tidy2["2021_gdp_num"] != '(D)') & (df_tidy2["2021_gdp_num"] != '(NA)')) | ((df_tidy2["2020_gdp_num"] != '(D)') & (df_tidy2["2020_gdp_num"] != '(NA)')) | (df_tidy2["2022_gdp_num"] == 'NaN')
```

```python
df_tidy2 = df_tidy2[mask2]
```

```python
df_tidy2["2022_gdp_num"] = pd.to_numeric(df_tidy2["2022_gdp_num"], errors="coerce") * 1000
```

```python
df_tidy2["2021_gdp_num"] = pd.to_numeric(df_tidy2["2021_gdp_num"], errors="coerce") * 1000
```

```python
df_tidy2["2020_gdp_num"] = pd.to_numeric(df_tidy2["2020_gdp_num"], errors="coerce") * 1000
```

```python
df_tidy2
```

```python
# Replace 'df' with whatever your dataframe variable is called
col = '2022_gdp_num'   # or the actual column name containing GDP values

# Show a few examples of non-numeric entries
mask_non_numeric = ~df_tidy2[col].astype(str).str.match(r"^[0-9,.\-]+$", na=False)
#bad_rows = df_tidy2.loc[mask_non_numeric, [col, 'state_county_fips_cd', 'naics_sector_desc']]
bad_rows = df_tidy2.loc[mask_non_numeric]

print(f"Non-numeric entries in column {col}: {bad_rows.shape[0]}")
print(bad_rows.head(20))  # show a few examples

# Count unique non-numeric values to see what they are
print("Unique non-numeric tokens:", df_tidy2.loc[mask_non_numeric, col].unique())
```
